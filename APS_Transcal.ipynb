```python
# biblioteca 

import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
k = 240
temperatura_ambiente =  298.15     # Temperatura 25°C 
temperatura_superficie = 373.15    # Temperatura da superficie a 100 °C
delta = temperatura_superficie - temperatura_ambiente
d = 0.240
l = 0.150
h = 20
```

# Questão I

### Item 1

```python
def extremidade_atleta(k,d,l,h,delta,temperatura_ambiente,b):
    m = math.sqrt((h*(math.pi*d))/(k*(((math.pi*(d**2))/4) + (2*math.pi*(d/2)*l))))

    t = ((math.cosh(m*(l-b)) + (h/(m*k)) * math.sinh(m*(l-b)))/(math.cosh(m*l) + (h/(m*k)) * math.sinh(m*l)))*delta + temperatura_ambiente

    return t
```

```python
distancia = np.arange(0.001,0.150,0.001)
temperatura = []

for i in range(len(distancia)):
    t = extremidade_atleta(k,d,l,h,delta,temperatura_ambiente,distancia[i])
    temperatura.append(t)

plt.plot(distancia, temperatura,'r')
plt.title('Temperatura ao longo do comprimento da aleta')
plt.xlabel('Distância em Metros (m)')
plt.ylabel('Temperatura em Kelvin (K)')
plt.grid()
plt.show()
```

### Item 2

```python
def transferencia_de_calor(k,d,l,h): 
    m = math.sqrt((h*(math.pi*d))/(k*(((math.pi*(d**2))/4) + (2*math.pi*(d/2)*l))))
    
    M = (math.cosh(m*l) + (h/(m*k)) * math.sinh(m*l))*math.sqrt(h*(math.pi*d)*k*(((math.pi*(d**2))/4) + (2*math.pi*(d/2)*l)))

    taxa = M*((math.sinh(m*l) + (h/(m*k)) * math.cosh(m*l))/math.cosh(m*l) + (h/(m*k)) * math.sinh(m*l))

    return taxa

print(transferencia_de_calor(k,d,l,h))
```

### Item 3

```python
def eficiência_aleta(k,d,l,h):
    m = math.sqrt((h*(math.pi*d))/(k*(((math.pi*(d**2))/4) + (2*math.pi*(d/2)*l))))
    theta_b = math.cosh(m*l) + (h/(m*k)) * math.sinh(m*l)
    taxa_transferencia_maxima = h*(((math.pi*(d**2))/4) + (2*math.pi*(d/2)*l))*theta_b
    taxa_trans = transferencia_de_calor(k,d,l,h)
    n = (taxa_trans/taxa_transferencia_maxima)

    return n

print('A eficiência da aleta é de {}'.format(eficiência_aleta(k,d,l,h)))
```
