```python
from math import pi, sqrt, cosh, sinh
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Dados do enunciado

k = 240
temperatura_ambiente =  298.15     # Temperatura 25°C 
temperatura_superficie = 373.15    # Temperatura da superficie a 100 °C
delta = temperatura_superficie - temperatura_ambiente
d = 0.240
l = 0.150
h = 20
```

```python
#Com isso, temos
P = pi*d
Atr = pi*(d**2)/4
m = sqrt(h*P/(k*Atr))

thetab = cosh(m*l) + (h/(m*k))*sinh(m*l)
M = sqrt(h*P*k*Atr*thetab)

As = Atr + 2*pi*(d/2)*l
Ab = Atr
```

# Questão I

### Item 1

```python
def temperatura_aleta(b):
    
    t = ((cosh(m*(l-b)) + (h/(m*k)) * sinh(m*(l-b))) / (thetab))*delta + temperatura_ambiente

    return t
```

```python
distancia = np.arange(0.001,0.150,0.001)
temperatura = []

for i in range(len(distancia)):
    t = temperatura_aleta(distancia[i])
    temperatura.append(t)

plt.plot(distancia, temperatura,'r')
plt.title('Temperatura ao longo do comprimento da aleta')
plt.xlabel('Distância em Metros (m)')
plt.ylabel('Temperatura em Kelvin (K)')
plt.grid()
plt.show()
```

```python
print(f'Na extremidade, T = {temperatura_aleta(0.15)}')
```

### Item 2

```python
def transferencia_de_calor(): 

    taxa = M*((sinh(m*l) + (h/(m*k)) * cosh(m*l))/(cosh(m*l) + (h/(m*k)) * sinh(m*l)))

    return taxa

print(transferencia_de_calor())
```

### Item 3

```python
def eficiência_aleta():
    taxa_transferencia_maxima = h*As*thetab
    taxa_trans = transferencia_de_calor()
    n = (taxa_trans/taxa_transferencia_maxima)

    return n

print('A eficiência da aleta é de {}'.format(eficiência_aleta()))
```

### Item 4

```python
def efetividade_aleta():
    ea = transferencia_de_calor() / (h * Ab * thetab)
    return ea

print('A efetividade da aleta é de {}'.format(efetividade_aleta()))
```
